```python
def firstStrat(maze):
    start = cell(0,0)
    queue = Queue()
    queue.put(start)
    #print(maze.dim)
    # next block of code is basically the bfs part
    visitedCells = []
    while queue.empty() == False:
        currCell = queue.get()
        if currCell.row == maze.dim - 1 and currCell.col == maze.dim - 1:
            break
        if currCell not in visitedCells:
            visitedCells.append(currCell)
            points = [(currCell.row, currCell.col + 1), (currCell.row + 1, currCell.col)]
            for p in points:
                if (p[0] >= 0 and p[1] >= 0 and p[0] < maze.dim and p[1] < maze.dim):
                    if maze.board[p[0]][p[1]] == 2 or maze.board[p[0]][p[1]] == -1:
                        continue
                    else:
                        neighbor = cell(p[0], p[1])
                        neighbor.prev = currCell
                        queue.put(neighbor)
    # load goal node into list
    path = [currCell]    
    # create list of cells that generate the path through the maze
    while not(currCell.prev == None):
        prev = currCell.prev
        path.append(prev)
        currCell = prev
#     for x in path:
#         print("i: {}".format(x.row))
#         print("j: {} \n".format(x.col))
#     for x in path:
#         print("({0}, {1})".format(x.row,x.col))
    path.reverse()
    return path
```

```python
def run_firstStrat(maze):
    m = maze
    path = firstStrat(m)
#     for x in path:
#         print("({0}, {1})".format(x.i,x.j))
    
    # using the initial path generated, iterate through the path, checking if one of the cells along the path catches fire
    for p in path:
        checki = p.row 
        checkj = p.col
        #print(checki)
        if maze[checki][checkj] == 2: # if the current cell is one of the cells on fire
            return "Agent died in fire"
        elif p.row == maze.dim - 1 and p.col == maze.dim - 1:
            return "Agent made it thru maze"
        else:
            m.iterateFire()
        
run_firstStrat(m)
```

```python
def generateCorners(listCorners):
    c1 = (listCorners[0][0] - 1, listCorners[0][1] - 1)
    c2 = (listCorners[1][0] - 1, listCorners[1][1] + 1)
    c3 = (listCorners[2][0] + 1, listCorners[2][1] - 1)
    c4 = (listCorners[3][0] + 1, listCorners[3][1] + 1)
    newCorners = [c1,c2,c3,c4]
    return newCorners

def calculateEuclideanDist(source, fire):
    row_source,col_source = source
    row_fire, col_fire = fire
    if row_source == row_fire:
        return abs(col_source - col_fire)
    elif col_source == col_fire:
        return abs(row_source - row_fire)
    else:
        rdiff = abs(row_source - row_fire)
        cdiff = abs(col_source - col_fire)
        return math.sqrt(rdiff**2 + cdiff**2)

def tryAgain(maze, cell):
    maxEuclideanDist = maze.dim * math.sqrt(2)
    row = cell.row
    col = cell.col
    if (maze[row][col] == 2):
        return 1 # this shouldn't actually be 1 but something else, 
    
    corners = generateCorners([(row,col), (row,col),(row,col),(row,col)])
    distCheck = 1
    fireList = []
    while distCheck < maze.dim:
        c1 = corners[0] # top left
        c2 = corners[1] # top right
        c3 = corners[2] # bottom left
        c4 = corners[3] # bottom right
        
        if c1[0]>=0: # if row is in bounds, so there should be section of code where it accounts for row not in bounds
            r = c1[0]
            for x in range(max(0,c1[1]), min(maze.dim-1, c2[1])): # c1 --> c2
                if maze[r][x] == 2:
                    # put some way to tell program to break out of big loop too, because closest fire been found in this grid
                    fireList.append((r,x))
        
        if c1[1]>=0: # if column is in bounds
            c = c1[1]
            for x in range(max(0,c1[0]), min(maze.dim-1, c3[0])): # c1 --> c3
                if maze[x][c] == 2:
                    fireList.append((x,c))
        
        if c2[1]>=0: # if column in bounds
            c = c2[1]
            for x in range(max(0,c2[0]), min(maze.dim-1, c4[0])): # c2 --> c4
                if maze[x][c] == 2:
                    fireList.append((x,c))
        
        if c3[0]<maze.dim:
            r = c3[0]
            for x in range(max(0,c3[1]), min(maze.dim-1, c4[1])): # c3 --> c4
                if maze[r][x] == 2:
                    fireList.append((r,x))
        
        
        corners = generateCorners([c1,c2,c3,c4])
        distCheck += 1
    minimum = maxEuclideanDist
    for fire in fireList:
        distance = calculateEuclideanDist((row,col), fire)
        if distance < minimum:
            minimum = distance
    return minimum
```
